## **1. Model Load**

```python
import os
import torch
from transformers import (
    AutoModelForCausalLM,
    AutoTokenizer,
    BitsAndBytesConfig,
    pipeline,
)
from peft import PeftModel, PeftConfig
from transformers import AutoModelForCausalLM
import warnings
warnings.filterwarnings('ignore')

BASE_ID = 'beomi/llama-2-ko-7b'
MODEL_ID = "Ash-Hun/WelSSiSKo_v3_llama-2-ko-base_text-generation"

# Setting Config
peft_config = PeftConfig.from_pretrained("Ash-Hun/WelSSiSKo_v3_llama-2-ko-base_text-generation")

bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type='nf4',
    bnb_4bit_compute_dtype="float16",
    bnb_4bit_use_double_quant=False
)


load_model = AutoModelForCausalLM.from_pretrained(BASE_ID, quantization_config=bnb_config, device_map={"":0})
load_model.config.use_cache = False
load_model.config.pretraining_tp = 1

base_tokenizer = AutoTokenizer.from_pretrained(BASE_ID, trust_remote_code=True)
base_tokenizer.pad_token = base_tokenizer.eos_token
base_tokenizer.padding_side = "right"

load_model = PeftModel.from_pretrained(load_model, MODEL_ID)

load_model.eval()
```

```python
def prompt_processing(user_input):
    return f'### Instruction:\n{user_input}\n\n### Response:'

def export_output(ouput_text):
    sep = ouput_text[0]['generated_text'].split('### Response:')[1].split('### Instruction')[0].split('## Instruction')[0].split('# Instruction')[0].split('Instruction')[0]
    sep = sep[1:] if sep[0] == '.' else sep
    sep = sep[:sep.find('.')+1] if '.' in sep else sep
    return sep

gen_pipe = pipeline(task="text-generation",
                model=load_model,
                tokenizer=base_tokenizer,
                max_length=500,
                do_sample=True,
                temperature=0.1,
                num_return_sequences=1,
                eos_token_id=base_tokenizer.eos_token_id,
                top_k=3,
                # top_p=0.3,
                repetition_penalty = 1.3,
                framework='pt'
                # early_stopping=True
)

def gen(prompt):
    result = gen_pipe(prompt_processing(prompt))
    print(export_output(result))
    return export_output(result)
```

## **2. Evaluate**

```python
from sentence_transformers import SentenceTransformer, util
import pandas as pd
import json
import re

model = SentenceTransformer("jhgan/ko-sroberta-multitask")

def cossim(s1, s2):
    print(s1, s2)
    s1 = re.sub(r'/^[ㄱ-ㅎ가-힣a-zA-Z0-9]*$/', "", s1)
    s2 = re.sub(r'/^[ㄱ-ㅎ가-힣a-zA-Z0-9]*$/', "", s2)
    s1_emb = model.encode(s1, convert_to_tensor=True)
    s2_emb = model.encode(s2, convert_to_tensor=True)
    cos_scores = util.pytorch_cos_sim(s1_emb, s2_emb)[0]
    cos_scores = cos_scores.cpu()
    return cos_scores

with open('eval_set.json') as f:
    js = json.loads(f.read()) ## json 라이브러리 이용

eval_score = []
for i in range(100):
    q = js[i]['Answer']
    q_pred = gen(js[i]['Question'])
    sim_score = cossim(q, q_pred)
    eval_score.append({'CosineScore':str(sim_score), 'Answers':q, 'Prediction':q_pred})

    print(f"> q : {q}")
    print(f"> q_pred : {q_pred}")
    print(f"> cosine similarity : {sim_score}")

with open("eval_score.json", "w", encoding="utf-8") as f:
    json.dump(eval_score, f, ensure_ascii=False, indent="\t")
```

```python
def transform(x):
    pattern = re.findall(r"tensor\(\[(.*?)\]\)", str(x))
    if pattern:
        return float(pattern[0])
    else:
        return x

# score_file = "./eval_score.json"
score_file = "./evaluate_result.json"
data = pd.read_json(score_file)
df = pd.DataFrame(data)
df['CosineScore'] = df['CosineScore'].apply(transform)

print(f"AVG Score : {df['CosineScore'].mean()}")
```
